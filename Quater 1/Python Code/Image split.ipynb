```python
import cv2

fruits = ['apple','banana','blueberry','cherry_tomato','kiwi','orange','pear','persimmon','raspberry']                                       # category

for fruit in fruits:                       
    
    read_folder_path = 'rawdata' + '/' + fruit

    for i in range(1,51):
        
        file_name = f'IMG_{i:04n}'                                       # f'IMG_{i:04n}' = 'IMG_{:04n}'.format(i)
        read_data_path = read_folder_path + '/' + f'{file_name}.jpg'     # f'/{file_name}.jpg' = '/{}.jpg'.format(file_name)  

        img = cv2.imread(read_data_path)

        # cv2.imread() -> takes an image as an input
        h, w, channels = img.shape

        # divide the image into left and right
        half = w//2

        # this will be the first column
        left_part = img[:, :half] 
        # [:,:half] means all the rows and
        # all the columns upto index half

        # this will be the second column
        right_part = img[:, half:]  
        # [:,half:] means al the rows and all
        # the columns from index half to the end
        
        # cv2.imwrite() function will save the image into your pc
        if i < 41:
            save_data_path_left = 'data/train' + '/' + fruit + '/' + f'{2*i-1}.jpg'
            save_data_path_right = 'data/train' + '/' + fruit + '/' f'{2*i}.jpg'
            cv2.imwrite(save_data_path_left, left_part)
            cv2.imwrite(save_data_path_right, right_part)
            cv2.waitKey(0)
        if 40 < i < 46:
            save_data_path_left = 'data/validation' + '/' + fruit + '/' + f'{2*i-1}.jpg'
            save_data_path_right = 'data/validation' + '/' + fruit + '/' f'{2*i}.jpg'
            cv2.imwrite(save_data_path_left, left_part)
            cv2.imwrite(save_data_path_right, right_part)
            cv2.waitKey(0)
        if 45 < i < 51:
            save_data_path_left = 'data/test' + '/' + fruit + '/' + f'{2*i-1}.jpg'
            save_data_path_right = 'data/test' + '/' + fruit + '/' f'{2*i}.jpg'
            cv2.imwrite(save_data_path_left, left_part)
            cv2.imwrite(save_data_path_right, right_part)
            cv2.waitKey(0)
```
